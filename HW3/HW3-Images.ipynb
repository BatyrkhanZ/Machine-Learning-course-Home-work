```python
import matplotlib.pyplot as plt
import matplotlib.image as pltimg
import numpy as np
```

```python
#Loading and showing image
%matplotlib inline
im = pltimg.imread('valley.jpg')
fig = plt.figure(figsize=(15,10))
plt.imshow(im)
```

```python
#Rotation of image
im_rotated90 = np.rot90(im,1)
im_rotated180 = np.rot90(im,2)
fig = plt.figure(figsize=(15,10))
plt.subplot(121)
plt.imshow(im_rotated90)
plt.subplot(122)
plt.imshow(im_rotated180)
```

```python
#combination of images

im2 = pltimg.imread('sea.jpg')
i,j,z = im.shape
im2_fixed = im2[0:i,0:j,0:z]
print(im.shape)
print(im2_fixed.shape)
im_comb1 = im+im2_fixed
fig = plt.figure(figsize=(15,10))
plt.subplot(221)
plt.imshow(im)
plt.subplot(222)
plt.imshow(im2_fixed)
plt.subplot(223)
plt.imshow(im_comb1)
```

```python
#Kernels
from scipy import ndimage
kern=np.ones((3,3))
filtered = ndimage.convolve(im[:,:,0], kern, mode='nearest', cval=0.0)
plt.imshow(filtered,cmap='gray')
```
