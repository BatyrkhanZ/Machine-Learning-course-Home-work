Task: Apply Linear Regression on Boston House dataset

```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
#Load dataset
Boston = datasets.load_boston()
data = Boston.data
target = Boston.target
```

```python
#Correlation coefficient
def estimate_coef(x, y): 
    n = np.size(x)

    # mean of x and y vector 
    m_x, m_y = np.mean(x), np.mean(y) 

    # calculating cross-deviation and deviation about x 
    SS_xy = np.sum(y*x) - np.sum(y*m_x)-np.sum(x*m_y) + n*m_y*m_x 
    print('k1',np.sum(y*x),' k2',np.sum(y*m_x), ' k3', np.sum(x*m_y), ' k4', n*m_y*m_x)
    SS_xx = np.sum(x*x) - n*m_x*m_x 

    # calculating regression coefficients 
    b_1 = SS_xy / SS_xx 
    b_0 = m_y - b_1*m_x 

    return(b_1, b_0)
```

```python
#Finding coefficients for all attributes to target
k = np.zeros(13)
b = np.zeros(13)

for i in range(0,13):
    k[i], b[i]= estimate_coef(target,data[:,i])

    

    
```

```python
#Scatter plot with regression
for i in range(0,13):
    fig = plt.figure(figsize=(10,20))
    plt.subplot(7,2,i+1)
    plt.scatter(target, data[:,i], alpha=0.5)
    reg_line = [(k[i]*)+b[i] for x in target]
    plt.plot(target, reg_line, color='red')  
    plt.title('Scatter plot and regression of attribute '+str(i+1))
    plt.xlabel('x')
    plt.ylabel('y')
    

    
plt.show()
```
